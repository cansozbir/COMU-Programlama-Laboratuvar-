### Size verilen bozukluk listesi ve para ustu degiskenleriyle en az sayida madeni para ile verilebilecek para ustunu hesaplayan kodu yaziniz

```python
bozukluklar = [1,5,10,21,25,50]
```

#### Kendi cozumum

```python
def para_ustu (miktar, bozukluklar):
    minimum_tane = miktar    #1 liklerle verilebilecek maksimum sayi miktarin kendisine esit
    hangi_bozukluklar = []
    
    for i in range (len(bozukluklar)):
        tane = 0
        temp_bozukluklar = []
        amount = miktar
        j=i
        while amount >= 0 and j>=0 :
            if amount - bozukluklar[j] >= 0:
                amount -= bozukluklar[j]
                tane += 1
                temp_bozukluklar.append(bozukluklar[j])
            else:
                j -= 1
        if tane < minimum_tane :
            minimum_tane = tane
            hangi_bozukluklar= temp_bozukluklar
    
    return minimum_tane,hangi_bozukluklar
```

```python
para_ustu(63, bozukluklar)
```

```python
para_ustu(64, bozukluklar)
```

```python
para_ustu(100, bozukluklar)
```

```python
para_ustu(15,bozukluklar)
```

```python
para_ustu(12421421,bozukluklar)
```

#### Derste gosterilen ornek

```python
def recMC(coinValueList,change,knownResults):
    minCoins = change
    
    if change in coinValueList:
        knownResults[change] = 1
        return 1
    elif knownResults[change] > 0:
        return knownResults[change]
    else :
        for i in [c for c in coinValueList if c<= change]:#bozuk para listesi olustur
            numCoins = 1 + recMC(coinValueList, change-i,knownResults)
            if numCoins < minCoins:
                minCoins = numCoins
                knownResults[change] = minCoins
    return minCoins
```

```python
recMC(bozukluklar, 40,[0]*41)
```

```python
recMC(bozukluklar,99,[0]*124)
```
