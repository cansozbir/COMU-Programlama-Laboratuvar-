```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = 'bird.jpg'
im_1 = plt.imread(file_name)
print(type(im_1))
print(im_1.shape)
print(im_1[0,0,0], im_1[0, 0, 1], im_1[0, 0, 2])
print(im_1[0][0])
```

```python
plt.imshow(im_1)
plt.show()
```

```python
def my_rot(img):
    my_new_image = np.zeros((img.shape[1], img.shape[0], 3), dtype=int)
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            my_new_image[col][row] = img[row][col]  # bunun kendisi de bir liste
            # my_new_image[col][row][:] = img[row][col][:]   bunlarin hepsi olur
            # my_new_image[col,row,:] = img[row,col,:]
    return my_new_image

    
im_2  = my_rot(im_1)
plt.subplot(1, 2, 1) # ucuncu 1 , 2 lik alanin 1.olanina alta ilk yazacagimi yaz demek
plt.imshow(im_1)
plt.subplot(1, 2, 2)
plt.imshow(im_2)
```

```python
def my_norm(my_v):  # vector uzunlugu
    return int((my_v[0]**2 + my_v[1]**2 + my_v[2]**2) ** 0.5)
```

```python
my_norm(im_1[0, 0, :])  # im_1[0][0][:] veya im_1[0][0]
```

```python
def my_convert_rgb_to_gray(img):
    my_new_image = np.copy(img)
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            my_new_image[row][col] = my_norm(my_new_image[row][col])
    return my_new_image

im_3 = my_convert_rgb_to_gray(im_1)

plt.subplot(1, 3, 1) # ucuncu 1 , 2 lik alanin 1.olanina alta ilk yazacagimi yaz demek
plt.imshow(im_1)
plt.subplot(1, 3, 2)
plt.imshow(im_2)
plt.subplot(1, 3, 3)
plt.imshow(im_3, cmap='gray')
```
